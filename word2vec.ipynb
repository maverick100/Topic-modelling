```python
from gensim.models import Word2Vec
import nltk
from sklearn.decomposition import PCA
from matplotlib import  pyplot as plt
```

```python
sent1="obama is the president of America"
sent2="winston churchill was the former head of United Stated"
sent3=""
```

```python
token=[]
for sent in ([sent1,sent2,sent3]):
    
    token.append(nltk.word_tokenize(sent))
          
          

```

```python
model=Word2Vec(token,min_count=1)
X=model[model.wv.vocab]
```

```python
%matplotlib notebook
pca=PCA(n_components=2)
result=pca.fit_transform(X)
plt.scatter(result[:,0],result[:,1])
words=list(model.wv.vocab)
for i,s in enumerate(words):
    plt.annotate(s,xy=(result[i,0],result[i,1]))
plt.show()
```
